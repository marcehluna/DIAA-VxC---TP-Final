```python
# Instalar ipyfilechooser si no est√° instalado
try:
    from ipyfilechooser import FileChooser
except ImportError:
    import sys
    !{sys.executable} -m pip install ipyfilechooser -q
    from ipyfilechooser import FileChooser

import ipywidgets as widgets
import os
import cv2

# Inicializar el selector de archivos
fc = FileChooser(os.getcwd())
fc.title = '<b>Selecciona la carpeta de im√°genes</b>'
fc.show_only_dirs = True

# Bot√≥n para cargar
btn_load = widgets.Button(
    description='Cargar Im√°genes',
    button_style='success',
    icon='check'
)

# Output para mostrar mensajes
out = widgets.Output()

# Variable global para almacenar las im√°genes
loaded_images = []

def load_images(b):
    global loaded_images
    folder_path = fc.selected_path
    image_extensions = ('.png', '.jpg', '.jpeg', '.gif', '.bmp', '.tiff', '.webp')
    loaded_images = []
    
    with out:
        out.clear_output()
        if not folder_path:
            print("‚ö†Ô∏è Por favor selecciona una carpeta.")
            return
        
        print(f"üìÇ Carpeta seleccionada: {folder_path}")
        if not os.path.exists(folder_path):
            print("‚ùå La carpeta no existe.")
            return

        print("‚è≥ Cargando im√°genes...")
        count = 0
        for filename in os.listdir(folder_path):
            file_path = os.path.join(folder_path, filename)
            if os.path.isfile(file_path) and filename.lower().endswith(image_extensions):
                img = cv2.imread(file_path, cv2.IMREAD_COLOR)
                if img is not None:
                    loaded_images.append(img)
                    count += 1
        
        print(f"‚úÖ Se cargaron {count} im√°genes correctamente en la variable 'loaded_images'.")

btn_load.on_click(load_images)
display(widgets.VBox([fc, btn_load, out]))
```
